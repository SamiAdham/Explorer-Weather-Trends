```python
#Import all libraries that we going to use it

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

#Read Dataframe from CSV file that been genrated by SQL
data = pd.read_csv("Result of Mecca Temp.csv")


# Create Function to calculate moving avarage (rolling)
def Rolling(myRange, myInput):
    result = myInput.rolling(window=myRange, on='city_avgtemp').mean().dropna()
    return result


# Calling function with High moving range
Chart_MAvg = Rolling(150, data)

# Draw results
plt.plot(Chart_MAvg['year'], Chart_MAvg['global_avgtemp'],label='global avg temp')
plt.plot(Chart_MAvg['year'], Chart_MAvg['city_avgtemp'],label='city avg temp')

plt.title('Explore Weather Trend')
plt.legend()
plt.ylabel('Temperature in C')
plt.xlabel('Years')
plt.show()
```

```python
# Calling function with Low moving range

Chart_MAvg = Rolling(50, data)

# Draw results
plt.plot(Chart_MAvg['year'], Chart_MAvg['global_avgtemp'],label='global avg temp')
plt.plot(Chart_MAvg['year'], Chart_MAvg['city_avgtemp'],label='city avg temp')

plt.title('Explore Weather Trend')
plt.legend()
plt.ylabel('Temperature in C')
plt.xlabel('Years')
plt.show()
```
